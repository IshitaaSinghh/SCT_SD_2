```python
import tkinter as tk
import random

secret_number = random.randint(1, 100)

def check_guess():
    try:
        user_guess = int(entry.get())
    except ValueError:
        result_label.config(text="Please enter a valid number!")
        return

    if user_guess < secret_number:
        result_label.config(text="Too low ðŸ˜”! Try again.")
    elif user_guess > secret_number:
        result_label.config(text="Too high ðŸ˜”! Try again.")
    else:
        result_label.config(text="Correct ðŸ¥³! You guessed the number ðŸŽ‰!")
 
window = tk.Tk()
window.title("Guess the Number Game")
window.geometry("300x200")

label = tk.Label(window, text="Guess a number between 1 and 100")
label.pack(pady=10)

entry = tk.Entry(window)
entry.pack(pady=5)

guess_button = tk.Button(window, text="Check Guess", command=check_guess)
guess_button.pack(pady=5)

result_label = tk.Label(window, text="")
result_label.pack(pady=10)

window.mainloop()
```
